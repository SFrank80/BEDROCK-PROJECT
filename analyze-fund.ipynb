```python
import boto3
import csv
import json
```

```python

bedrock_client = boto3.client('bedrock-runtime')
```

```python
jan_holdings = "jan_holdings.csv"
```

```python
with open(jan_holdings, mode="r", newline="", encoding="utf-8") as file:
    csv_reader = csv.reader(file)

    formatted_jan_holdings = ""

    for row in csv_reader:
        formatted_jan_holdings += " | ".join(row) + "\n"
```

```python
formatted_jan_holdings
```

```python
feb_holdings = "feb_holdings.csv"
```

```python
with open(feb_holdings, mode="r", newline="", encoding="utf-8") as file:
    csv_reader = csv.reader(file)

    formatted_feb_holdings = ""

    for row in csv_reader:
        formatted_feb_holdings += " | ".join(row) + "\n"
```

```python
formatted_feb_holdings
```

```python
prompt_config = {
    "anthropic_version": "bedrock-2023-05-31",
    "temperature": 1,
    "top_p": 0.999,
    "top_k": 250,
    "max_tokens": 2000, 
    "stop_sequences": ["\n\nHuman:"], 
    "messages": [
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "I have provided you with two documents named formatted_jan_holding and formatted_feb_holding. Please compare these two documents and create a list of the top 30 holdings whose share count has changed between January and February. For each holding, please only provide the holdings name, ticker symbol and the difference in shares between January and February.:"},
                {"type": "text", "text": formatted_jan_holdings}, 
                {"type": "text", "text": formatted_feb_holdings},
            ],
        }
    ],
}
```

```python
body = json.dumps(prompt_config)
contentType = "application/json"
accept = "application/json"
modelId = "anthropic.claude-3-haiku-20240307-v1:0"
```

```python
response = bedrock_client.invoke_model(
    body=body, 
    contentType=contentType,
    accept=accept,
    modelId=modelId,  
)
```

```python
response_body = json.loads(response.get("body").read())

results = response_body.get("content")[0].get("text")
```

```python
results
```

```python
filename = "fiinding.txt"
```

```python
with open(filename, "w") as file:
    file.write(results)
```
